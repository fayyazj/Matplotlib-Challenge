```python
# Dependencies and Setup
# %matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
Mouse_Data_DF = pd.read_csv(mouse_drug_data_to_load)
Clinical_Trial_Data_DF = pd.read_csv(clinical_trial_data_to_load)
```

```python
#Look at Mouse_Data
Mouse_Data_DF.head()
```

```python
#Look at Clinical Trial Data
Clinical_Trial_Data_DF.head()
```

```python
# Combine the data into a single dataset
Complete_Drug_Data_DF = pd.merge(Mouse_Data_DF, Clinical_Trial_Data_DF, on='Mouse ID', how='left')

# Display the data table for preview
Complete_Drug_Data_DF.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint, Convert to DataFrame
Mean_Tumor = Complete_Drug_Data_DF.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)']

#Convert to DataFrame
Mean_Tumor_DF = pd.DataFrame(Mean_Tumor)

Mean_Tumor_DF = Mean_Tumor_DF.reset_index()

# Preview DataFrame
Mean_Tumor_DF.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
Sem_Tumor = Complete_Drug_Data_DF.groupby(["Drug", "Timepoint"]).sem()['Tumor Volume (mm3)']

# Convert to DataFrame
Sem_Tumor_DF = pd.DataFrame(Sem_Tumor)

Sem_Tumor_DF = Sem_Tumor_DF.reset_index()

# Preview DataFrame
Sem_Tumor_DF.head()
```

```python
# Minor Data Munging to Re-Format the Mean_Tumor_DF Data Frame
Mean_Tumor_DF = Mean_Tumor_DF.pivot_table(values = "Tumor Volume (mm3)", columns = "Drug", index = "Timepoint")

# Preview that Reformatting worked
Mean_Tumor_DF.head()
```

```python
# Minor Data Munging to Re-Format the Sem_Tumor_DF Data Frame
Sem_Tumor_DF = Sem_Tumor_DF.pivot_table(values = "Tumor Volume (mm3)", columns = "Drug", index = "Timepoint")

# Preview that Reformatting worked
Sem_Tumor_DF.head()
```

```python
# Generate the Plot (with Error Bars)

fig, ax = plt.subplots()

ax.errorbar(Mean_Tumor_DF.index, Mean_Tumor_DF["Capomulin"],yerr=Sem_Tumor_DF["Capomulin"], fmt="d:c",label="Capomulin")
ax.errorbar(Mean_Tumor_DF.index, Mean_Tumor_DF["Infubinol"], yerr=Sem_Tumor_DF["Infubinol"], fmt="^:m", label="Infubinol")
ax.errorbar(Mean_Tumor_DF.index, Mean_Tumor_DF["Ketapril"], yerr=Sem_Tumor_DF["Ketapril"], fmt="s:r", label="Ketapril")
ax.errorbar(Mean_Tumor_DF.index, Mean_Tumor_DF["Placebo"], yerr=Sem_Tumor_DF["Placebo"], fmt="o:g", label="Placebo")                                                    

plt.title("Tumor Response to Treatment")                        
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
plt.grid(axis="y")

plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Tumor Response to Treatment.png")

#To show figure use code below, however the plt.savefig wouldn't work if I did that, thus it is commented out
#plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
MetMean_Site_Data = Complete_Drug_Data_DF.groupby(["Drug", "Timepoint"]).mean()['Metastatic Sites']

# Convert to DataFrame
MetMean_Site_Data_DF = pd.DataFrame(MetMean_Site_Data)

# Preview DataFrame
MetMean_Site_Data_DF.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
MetSem_Site_Data = Complete_Drug_Data_DF.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
MetSem_Site_Data_DF = pd.DataFrame(MetSem_Site_Data)

# Preview DataFrame
MetSem_Site_Data_DF.head()
```

```python
 # Minor Data Munging to Re-Format the MetMean_Site_Data_DF Data Frame
MetMean_Site_Data_DF = MetMean_Site_Data_DF.pivot_table(values = "Metastatic Sites", columns = "Drug", index = "Timepoint")

# Preview that Reformatting worked
MetMean_Site_Data_DF.head()
```

```python
 # Minor Data Munging to Re-Format the Met_Sem_DF Data Frame
MetSem_Site_Data_DF = MetSem_Site_Data_DF.pivot_table(values = "Metastatic Sites", columns = "Drug", index = "Timepoint")

# Preview that Reformatting worked
MetSem_Site_Data_DF.head()
```

```python
 # Generate the Plot (with Error Bars)

fig, ax = plt.subplots()

ax.errorbar(MetMean_Site_Data_DF.index, MetMean_Site_Data_DF["Capomulin"],yerr=MetSem_Site_Data_DF["Capomulin"], fmt="d:c",label="Capomulin")
ax.errorbar(MetMean_Site_Data_DF.index, MetMean_Site_Data_DF["Infubinol"], yerr=MetSem_Site_Data_DF["Infubinol"], fmt="^:m", label="Infubinol")
ax.errorbar(MetMean_Site_Data_DF.index, MetMean_Site_Data_DF["Ketapril"], yerr=MetSem_Site_Data_DF["Ketapril"], fmt="s:r", label="Ketapril")
ax.errorbar(MetMean_Site_Data_DF.index, MetMean_Site_Data_DF["Placebo"], yerr=MetSem_Site_Data_DF["Placebo"], fmt="o:g", label="Placebo")                                                    

plt.title("Metastatic Spread During Treatment")                        
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Met Sites")
plt.grid(axis="y")

plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Metastatic Spread During Treatment")


#To show figure use code below, however the plt.savefig wouldn't work if I did that, thus it is commented out
#plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
Mouse_Count = Complete_Drug_Data_DF.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]

# Convert to DataFrame
Mouse_Count_DF = pd.DataFrame(Mouse_Count)
Mouse_Count_DF = Mouse_Count_DF.reset_index()
Mouse_Count_DF = Mouse_Count_DF.rename(columns= {"Mouse ID": "Mouse Count"})

# Preview DataFrame
Mouse_Count_DF.head()
```

```python
 # Minor Data Munging to Re-Format the Mouse_Count_DF Data Frame
Mouse_Count_DF = Mouse_Count_DF.pivot_table(values = "Mouse Count", columns = "Drug", index = "Timepoint")

# Preview that Reformatting worked
Mouse_Count_DF
```

```python
# Generate the Plot (Accounting for percentages)
fig, ax = plt.subplots()

ax.errorbar(Mouse_Count_DF.index, Mouse_Count_DF["Capomulin"]/Mouse_Count_DF["Capomulin"].iloc[0] * 100, fmt="d:c",label="Capomulin")
ax.errorbar(Mouse_Count_DF.index, Mouse_Count_DF["Infubinol"]/Mouse_Count_DF["Infubinol"].iloc[0] * 100, fmt="^:m", label="Infubinol")
ax.errorbar(Mouse_Count_DF.index, Mouse_Count_DF["Ketapril"]/Mouse_Count_DF["Ketapril"].iloc[0] * 100, fmt="s:r", label="Ketapril")
ax.errorbar(Mouse_Count_DF.index, Mouse_Count_DF["Placebo"]/Mouse_Count_DF["Placebo"].iloc[0] * 100, fmt="o:g", label="Placebo")                                                    

plt.title("Survival During Treatment")                        
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate (%)")
plt.grid(axis="both")
plt.ylim(33,105)
plt.legend(loc="lower left", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("Survival During Treatment")

#To show figure use code below, however the plt.savefig wouldn't work if I did that, thus it is commented out
#plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
Tumor_Percent_Change = 100 * (Mean_Tumor_DF.iloc[-1] - Mean_Tumor_DF.loc[0])/Mean_Tumor_DF.iloc[0]

# Display the data to confirm
Tumor_Percent_Change
```

```python
#Initialize Graph List & Graph
Graph_List = ['Capomulin','Infubinol','Ketapril','Placebo']
fig = plt.figure()
ax = fig.add_subplot()
x = -0.15

# Splice the data between passing and failing drugs/plot points
for drug in Graph_List:
    
    # Conditional formatting/splicing for failing drugs
    if Tumor_Percent_Change[drug] > 0:
        barcolor = 'red'
        y = 5
        
    # Conditional formatting/splicing for passing drugs
    else:
        barcolor = 'green'
        y = -7
    
    #Plot bar chart
    ax.bar(drug, Tumor_Percent_Change[drug], width=1, color=barcolor)
    
    #labeling bar with percentages
    ax.text(x, y, '{:.0f}%'.format(Tumor_Percent_Change[drug]), color='white')
    x += 1
    
# Orient widths. Add labels, tick marks, etc. 
ax.grid()
ax.set_autoscaley_on(False)
ax.set_ylim(-30, 70)
ax.set_xticks([0.5, 1.5 ,2.5 ,3.5])
ax.set_xticklabels(Graph_List)
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')

# Save the Figure
plt.savefig("Tumor Change Over 45 Day Treatment")

#To show figure use code below, however the plt.savefig wouldn't work if I did that, thus it is commented out
#plt.show()
```
